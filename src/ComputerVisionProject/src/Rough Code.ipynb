```python
import os
from PIL import Image
path=os.getcwd()
path=path+'\\Output\\'
for file in os.listdir(path):
    ex=file[-3:]
    if(ex=='png'):
        filepath=path+file
        img=Image.open(filepath)
        img=img.convert('RGB')
        name=''
        for i in range(len(file)-3):
            name+=file[i]
        name=name+'jpg'
        out='Output//'
        print(name)
        name=out+name
        img.save(name)
        os.remove(filepath)
```
